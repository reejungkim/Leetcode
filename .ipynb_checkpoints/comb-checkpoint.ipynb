```python
path = "형용사.csv"
```

```python
import numpy as np
import pandas as pd

list_of_list = pd.read_csv(path, header=0).values.tolist()
flattened = [val.replace(' ','') for sublist in list_of_list for val in sublist if str(val)!='nan']
adj = np.unique(flattened)
```

```python
list_of_list = pd.read_csv("동물.csv").values.tolist()
flattened = [val.replace(' ','').replace('.','') for sublist in list_of_list for val in sublist if str(val)!='nan']
nouns = flattened
```

```python
len(adj), len(nouns)
```

```python
number = range(1,100)
```

```python
from itertools import product 

items = [adj, nouns, number] 
list_id = list(product(*items))
```

```python
Ids = []
for id in list_id:
    Ids.append( '_'.join(map(str, id)) )
```
