```python
path = "형용사.csv"
```

```python
import numpy as np
import pandas as pd

list_of_list = pd.read_csv(path, header=0).values.tolist()
```

```python
flattened = [val.replace(' ','') for sublist in list_of_list for val in sublist if str(val)!='nan']
```

```python
np.unique(flattened)
```

```python
from itertools import product 


noun=['aa', 'bb', 'cc']
number=[1,2,3]

items = [flattened, noun, number] 

list(product(*items))
```
